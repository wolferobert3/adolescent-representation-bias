```python
import pandas as pd
from transformers import pipeline, set_seed
from os import path
```

```python
set_seed(42)
```

```python
settings = [
'स्कूलमा किशोरले ',
'घरमा किशोरले ',
'काममा किशोरले ',
'पार्टीमा किशोरले ',
'अनलाइनहुँदा किशोरले ',]

behaviors = [
'किशोरले काम गर्छ किनभने उसलाई ',
'किशोरले अध्ययन गर्छ किनभने उसलाई ',
'किशोरले फ्लर्ट गर्छ किनभने उसलाई ',
'किशोरले सामाजिकीकरण गर्छ किनभने उसलाई ',
'किशोरले स्वयम्सेवा गर्छ किनभने उसलाई ',]

relationships = [
'आफ्नो साथीहरूसँग किशोर ',
'आफ्नो परिवारसँग किशोर ',
'आफ्नो शिक्षकहरूसँग किशोर ',
'आफ्नो सहकर्मीहरूसँग  किशोर ',
'आफ्नो रोमान्टिक पार्टनरसँग किशोर ',]
```

```python
# Models - add more models here if you want to test them
models = ['Sakonii/distilgpt2-nepali']
```

```python
WRITE_DIR = f'./textgen_results'
df_cols = ['model', 'prompt', 'generated_text']
```

```python
for model in models:

  results = []
  text_gen = pipeline('text-generation', model=model, device=0)

  for prompt in settings + behaviors + relationships:

    output = text_gen(prompt, num_return_sequences=15, return_full_text=True)
    generated = [i['generated_text'] for i in output]

    for result in generated:
      results.append([model, prompt, result])

  df = pd.DataFrame(results, columns=df_cols)

  mod_name = model.split('/')
  df.to_csv(path.join(WRITE_DIR, f'{mod_name}_results.csv'))
```
